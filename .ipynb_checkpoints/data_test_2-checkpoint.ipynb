```python
import pandas as pd
import numpy as np
from datetime import datetime, date


file = '/Users/tjp135/Desktop/311_Service_Requests_from_2010_to_Present.csv' 
population = '/Users/tjp135/Desktop/2010+Census+Population+By+Zipcode+(ZCTA).csv'
row_count = 1000
datafile = pd.read_csv(file, usecols = ["Created Date", "Complaint Type", "Incident Zip", "Borough"], dtype = 'str', chunksize=row_count)
dataAll = pd.DataFrame()

for chunk in datafile: 
    data =  chunk.loc[(chunk['Created Date'].str.contains('2017'))]
#   bronx = chunk.loc[(chunk['Created Date'].str.contains('2017')) & (chunk['Borough'].str.contains('BRONX'))]

    
    if not data.empty:
        dataAll = pd.concat([dataAll, data], ignore_index=True)
    

        
dataAll['Complaint Type'].value_counts().nlargest(10)

   
```
